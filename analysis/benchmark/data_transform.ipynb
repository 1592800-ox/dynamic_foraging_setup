```python
from os import listdir, mkdir
from os.path import join as pjoin
from os.path import isdir

import pandas as pd
import numpy as np
from datetime import datetime
```

```python
for mouse_code in listdir('original_data'):
    dir_path = pjoin('original_data', mouse_code)
    for version in listdir(dir_path):
        v = version.split('_')[2]
        print(v)
        version_path = pjoin(dir_path, version)
        for session in listdir(version_path):
            trial_indices = []
            leftP = []
            rightP = []
            choices = []
            rewarded = []

            trial_index = 0
            
            # Gamble arm is left
            session_path = pjoin(version_path, session)
            date = '.'.join((session.split('-')[-1].split('.')[:-1]))
            if not isdir(pjoin('data', mouse_code)):
                mkdir(pjoin('data', mouse_code))
            des_file_path = pjoin('data', mouse_code, '%s-%s' % (date, v))
            print(des_file_path)
            with open(session_path, 'r', encoding='utf-16le') as f:
                for line in f:
                    if len(line.split(';')) < 20:
                        continue
                    data = line.split(';')
                    reward_choice = data[13]
                    if 'right' in reward_choice:
                        choices.append(1)
                    elif 'left' in reward_choice:
                        choices.append(0)
                    else:
                        choices.append(-1)

                    if 'rewarded' in reward_choice:
                        rewarded.append(1)
                    elif 'NOreward' in reward_choice:
                        rewarded.append(0)
                    else:
                        rewarded.append(-1)

                    left_prob = float(data[17][-3:].replace(',', '.'))
                    right_prob = float(data[18][-3:].replace(',', '.'))

                    leftP.append(left_prob)
                    rightP.append(right_prob)

                    trial_indices.append(trial_index)
                    trial_index += 1
            
            session_data = pd.DataFrame({'trial_indices': trial_indices, 'leftP': leftP, 'rightP': rightP, 'choices': choices, 'rewarded': rewarded})

            session_data.to_csv(des_file_path + '.csv', index=False)
```
